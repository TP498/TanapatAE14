## Run to import data from Firebase

```
#@title Firebase information
project_name = "tanapat-ae14-sensor" #@param {type:"string"}
data_path = "Drive to home 1" #@param {type:"string"}
import urllib.request
import pandas as pd
import json
url = 'https://tanapat-ae14-sensor.firebaseio.com/Drive%20to%20home%201.json'
resp = urllib.request.urlopen(url)
json_data = json.load(resp)
data = {'timestamp':[], 'latitude':[], 'longitude':[], 'x_avg':[], 'x_stdev':[], 'x_min':[], 'x_max':[], 'y_avg':[], 'y_stdev':[], 'y_min':[], 'y_max':[], 'z_avg':[], 'z_stdev':[], 'z_min':[], 'z_max':[]}
for record in json_data.values():
  sz = len(record['latitude'])
  data['timestamp'].extend( sz*[record['timestamp']] )
  data['latitude'].extend( [v for v in record['latitude']] )
  data['longitude'].extend( [v for v in record['longitude']] )  
  data['x_avg'].extend( [v for v in record['motionXAvg']] )
  data['x_stdev'].extend( [v for v in record['motionXStdev']] )
  data['x_min'].extend( [v for v in record['motionXMin']] )
  data['x_max'].extend( [v for v in record['motionXMax']] )
  data['y_avg'].extend( [v for v in record['motionYAvg']] )
  data['y_stdev'].extend( [v for v in record['motionYStdev']] )
  data['y_min'].extend( [v for v in record['motionYMin']] )
  data['y_max'].extend( [v for v in record['motionYMax']] )
  data['z_avg'].extend( [v for v in record['motionZAvg']] )
  data['z_stdev'].extend( [v for v in record['motionZStdev']] )
  data['z_min'].extend( [v for v in record['motionZMin']] )
  data['z_max'].extend( [v for v in record['motionZMax']] )  
df = pd.DataFrame.from_dict(data)
df
```

##Run to visualize data

```
#@title Data visualize
line1 = "z_avg" #@param ["x_avg", "x_stdev", "x_min", "x_max", "y_avg", "y_stdev", "y_min", "y_max", "z_avg", "z_stdev", "z_min", "z_max"]
line2 = "z_min" #@param ["x_avg", "x_stdev", "x_min", "x_max", "y_avg", "y_stdev", "y_min", "y_max", "z_avg", "z_stdev", "z_min", "z_max"]
line3 = "z_max" #@param ["x_avg", "x_stdev", "x_min", "x_max", "y_avg", "y_stdev", "y_min", "y_max", "z_avg", "z_stdev", "z_min", "z_max"]

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = [15, 7.5]
plt.plot( df[line1] )
plt.plot( df[line2] )
plt.plot( df[line3] )
plt.legend([line1, line2, line3])
plt.show()
```
